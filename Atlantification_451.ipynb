```python
#Importing stuff for graphs
import numpy as np
import xarray as xr
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import netCDF4 
import dask 
import regionmask 
import scipy
```

```python
#JJA Dataset
djf1 = xr.open_dataset("/Users/katrishka_kipp/Downloads/JJA_DataSet/202512_am2_v06r00.nc")
djf2 = xr.open_dataset("/Users/katrishka_kipp/Downloads/JJA_DataSet/202601_am2_v06r00.nc")
djf3 = xr.open_dataset("/Users/katrishka_kipp/Downloads/JJA_DataSet/202602_am2_v06r00.nc")
```

```python
#Picking variable. SIC = Sea Ice Concentration!
SIC1 = djf1['cdr_seaice_conc_monthly']
SIC2 = djf2['cdr_seaice_conc_monthly']
SIC3 = djf3['cdr_seaice_conc_monthly']
```

```python
#Slicing and setting time
barents_SIC1=SIC1.isel(x=slice(50,300), 
                       y=slice(50,300))
barents_SIC2=SIC2.isel(x=slice(50,300), 
                       y=slice(50,300))
barents_SIC3=SIC3.isel(x=slice(50,300), 
                       y=slice(50,300))

barents_SIC1=barents_SIC1.isel(time=0)
barents_SIC2=barents_SIC2.isel(time=0)
barents_SIC3=barents_SIC3.isel(time=0)
```

```python
#Create maps
proj = ccrs.NorthPolarStereo()

fig, axes = plt.subplots(
    1, 3,
    figsize=(18, 6),
    subplot_kw={"projection": proj}
)

print(type(axes[0]))
```

```python
#Plotting!
ds_djf=[barents_SIC1, barents_SIC2, barents_SIC3]
titles=['December', 'January', 'February']

fig, axes=plt.subplots(1, 3, figsize=(18,6))

for ax, data, title in zip(axes, ds_djf, titles):
   
    ax.set_title('DJF Sea Ice Concentration', loc = 'center')
    ax.set_xlabel("Projected Coordinate (m)")
    ax.set_ylabel("Projected Coordinate (m)")

    im = ax.pcolormesh(
     data["x"],
     data["y"],
     data,
     cmap="Blues",
     shading = 'auto',
     
   )
    
cbar = fig.colorbar(
    im, 
    ax=axes,
    orientation="horizontal",
     pad = 0.05
)
cbar.set_label("Sea Ice Concentration (%*100)", fontsize=8)

plt.show()
```
